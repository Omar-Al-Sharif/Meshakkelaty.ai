```python
def find_repeated_lines_with_counts_and_locations(file_path):
    """Finds and prints repeated lines with counts and line numbers."""
    with open(file_path, "r+") as file:
        lines_seen = {}  # Store line counts and first occurrences
        repeated_lines = {}  # Store repeated lines with counts and occurrences

        line_number = 0
        for line in file:
            line_number += 1
            stripped_line = line.strip()
            if stripped_line in lines_seen:
                count, first_occurrence, _ = repeated_lines.get(stripped_line, (0, 0, 0))  # Use get() with default tuple
                repeated_lines[stripped_line] = (
                    count + 1,  # Increment count
                    first_occurrence,
                    line_number,  # Update last occurrence
                )
            else:
                lines_seen[stripped_line] = line_number


    if repeated_lines:
        print("Repeated lines with counts and occurrences:")
        for line, (count, first_occurrence, last_occurrence) in repeated_lines.items():
            print(f"- Line '{line}' repeated {count} times:")
            print(f"   First occurrence: line {first_occurrence}")
            print(f"   Last occurrence: line {last_occurrence}")
    else:
        print("No repeated lines found in the file.")

# Example usage
file_path = "total_train.txt"  # Replace with your actual file path
find_repeated_lines_with_counts_and_locations(file_path)
```

```python
def remove_repeated_lines_and_write_unique(file_path, output_path):
    """Removes repeated lines from a file and writes unique lines to a new file."""
    unique_lines = set()  # Use a set to efficiently track unique lines

    with open(file_path, "r") as file, open(output_path, "w") as output_file:
        for line in file:
            stripped_line = line.strip()
            if stripped_line not in unique_lines:
                unique_lines.add(stripped_line)
                output_file.write(line)  # Write unique lines to the output file


file_path = "total_train.txt"
output_path = "final_dataset.txt"  
remove_repeated_lines_and_write_unique(file_path,output_path)
```
